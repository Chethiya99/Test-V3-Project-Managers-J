```python
import pandas as pd

# Load the two CSV files into DataFrames
file1 = 'merchants_sg_emails.csv'  # Replace with the path to your first CSV file
file2 = 'ModifiedSGRestaurants.csv'  # Replace with the path to your second CSV file

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Append the DataFrames
appended_df = pd.concat([df1, df2], ignore_index=True)

# Save the appended DataFrame to a new CSV file
output_file = 'merchants_sg_emails_updated.csv'  # Replace with the desired output file path
appended_df.to_csv(output_file, index=False)

print(f"Appended file saved to {output_file}")
```

```python
# Step 1: CSV file path
csv_file = 'real data.csv'  # Replace with your CSV file name or path

# Step 2: Read CSV file using pandas
df = pd.read_csv(csv_file)

df.head()
```

```python
df['Region'] = 'Singapore'
```

```python
df.head(2)
```

```python
# Step 1: Clean column names by replacing spaces with underscores
df.columns = df.columns.str.replace(' ', '_').str.replace(r'\W', '', regex=True)
```

```python
df['Restaurant_Name'].nunique()
```

```python
# Extract the first word from the 'Restaurant_Name' column and append '@gmail.com'
df['Email'] = df['Restaurant_Name'].str.split().str[0] + '@gmail.com'

# Display the updated dataframe
df.head()
```

```python
df.to_csv('Merchants updated.csv', index=False)
```

```python
# Step 3: Connect to SQLite Database (this creates the file if it doesn't exist)
db_name = 'merchant_data_singapore.db'  # SQLite database name
conn = sqlite3.connect(db_name)
cursor = conn.cursor()

# Step 4: Create the table (only run this once; otherwise, comment this part)
create_table_query = """
CREATE TABLE IF NOT EXISTS merchants (
    Position INTEGER,
    Promotion TEXT,
    Discount_Offer TEXT,
    Restaurant_Name TEXT,
    Rating REAL,
    Cuisines TEXT,
    Cost_for_Two TEXT,
    Location TEXT,
    Distance TEXT,
    Restaurant_URL TEXT,
    Region TEXT,
    Image_URL TEXT,
    Zomato_Gold_Image TEXT,
    Email TEXT
);
"""
cursor.execute(create_table_query)

# Step 5: Insert data into the table
# Use pandas to_sql function to insert the data into the database
df.to_sql('merchants', conn, if_exists='append', index=False)

# Step 6: Verify data insertion (Optional)
print("Data inserted successfully!")
query = "SELECT * FROM merchants LIMIT 5;"
for row in cursor.execute(query):
    print(row)

# Step 7: Close connection
conn.commit()
conn.close()
```
